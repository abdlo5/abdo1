```python
# =========================================
# Work From Home Burnout Dataset â€” Overview
# =========================================




# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt


#  Load Dataset
# If your CSV is in the repo ROOT:
df = pd.read_csv("../work_from_home_burnout_dataset.csv")

# If inside a data folder, use this instead:
# df = pd.read_csv("../data/work_from_home_burnout_dataset.csv")


#  Display First Rows
print("First 5 Rows:")
display(df.head())


#  Dataset Shape
print("Dataset Shape (Rows, Columns):", df.shape)


#  Column Names
print("\nColumns in the dataset:")
print(df.columns.tolist())


#  Data Types
print("\nData Types:")
display(df.dtypes)


#  Missing Values
missing_values = df.isnull().sum()
print("\nMissing Values per Column:")
display(missing_values)


#  Summary Statistics
print("\nSummary Statistics:")
display(df.describe())


#  Target Distribution (IMPORTANT for ML)
print("\nBurnout Risk Distribution:")
display(df["burnout_risk"].value_counts())


# Visualization (Makes notebook look more professional)
df["burnout_risk"].value_counts().plot(kind='bar')

plt.title("Burnout Risk Distribution")
plt.xlabel("Burnout Level")
plt.ylabel("Count")

plt.show()


#  Conclusion
print("""
Conclusion:
The dataset was successfully loaded and explored.
Basic statistics and class distribution were analyzed,
preparing the dataset for machine learning modeling.
""")
```
