This notebook loads the dataset and displays basic information (shape, columns, dtypes, missing values, summary stats, and class balance).

```python
import pandas as pd

DATA_PATH = r"/mnt/data/work_from_home_burnout_dataset.csv"
df = pd.read_csv(DATA_PATH)

df.head()
```

```python
# Shape (rows, columns)
df.shape
```

```python
# Column names
df.columns
```

```python
# Data types
df.dtypes
```

```python
# Missing values (count and %)
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(2)
pd.DataFrame({"missing_count": missing, "missing_pct": missing_pct}).sort_values("missing_count", ascending=False)
```

```python
# Numeric summary statistics
df.describe().T
```

```python
# Class balance for burnout_risk (label)
df["burnout_risk"].value_counts()
```

```python
# Simple plot: burnout_risk distribution
import matplotlib.pyplot as plt

df["burnout_risk"].value_counts().plot(kind="bar")
plt.xlabel("burnout_risk")
plt.ylabel("count")
plt.title("Burnout Risk Class Distribution")
plt.show()
```
